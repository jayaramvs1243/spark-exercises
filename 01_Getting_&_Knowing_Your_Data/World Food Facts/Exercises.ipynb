### Step 3. Use the tsv file and assign it to a dataframe called food

```python
from pyspark.sql import SparkSession

spark_context = SparkSession.builder.appName('Spark Exercises').getOrCreate()

df_food \
    = spark_context \
        .read \
        .csv(path=FILE_PATH, sep='\t', header=True, inferSchema=True)
df_food.show(5)
```

### Step 4. See the first 5 entries

```python
df_food.limit(5).show()
```

### Step 5. What is the number of observations in the dataset?

```python
from pyspark.sql import functions as F

df_food.select(F.count('*').alias('number_of_observations')).show()
```

### Step 10. How is the dataset indexed?

```python
df_food = df_food.withColumn('product_index', F.monotonically_increasing_id())
df_food.show(10)
```

### Step 11. What is the product name of the 19th observation?

```python
COLUMN_INDEX = 19

df_food \
    .filter((df_food.product_index == (COLUMN_INDEX - 1))) \
    .select(df_food.product_name) \
    .show()

df_food \
    .offset(COLUMN_INDEX - 1) \
    .limit(1) \
    .select(df_food.product_name) \
    .show()
```
